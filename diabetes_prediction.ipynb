```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.datasets import load_diabetes
from sklearn.datasets import fetch_openml

# Pima Indians Diabetes Dataset
diabetes = fetch_openml(name='diabetes', version=1, as_frame=True)
df = diabetes.frame

# Target fix karo
df['target'] = (df['class'] == 'tested_positive').astype(int)
df = df.drop('class', axis=1)
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()

print("Dataset shape:", df.shape)
print("\nPehli 5 rows:")
df.head()
```

```python
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDiabetes Cases:")
print(df['target'].value_counts())
print(f"\n0 = Healthy: {df['target'].value_counts()[0]}")
print(f"1 = Diabetic: {df['target'].value_counts()[1]}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Diabetes count
df['target'].value_counts().plot(
    kind='bar', ax=axes[0,0],
    color=['#2ecc71','#e74c3c']
)
axes[0,0].set_title('Diabetes Cases\n(0=Healthy, 1=Diabetic)')

# Age distribution
sns.histplot(df['age'], bins=20,
             ax=axes[0,1], color='#3498db')
axes[0,1].set_title('Age Distribution')

# Glucose vs Diabetes
sns.boxplot(x='target', y='plas', data=df,
            palette=['#2ecc71','#e74c3c'], ax=axes[1,0])
axes[1,0].set_title('Glucose Level vs Diabetes')

# BMI vs Diabetes
sns.boxplot(x='target', y='mass', data=df,
            palette=['#2ecc71','#e74c3c'], ax=axes[1,1])
axes[1,1].set_title('BMI vs Diabetes')

plt.tight_layout()
plt.show()
```

```python
X = df.drop('target', axis=1)
y = df['target']

# Data scale karo - Advanced step!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# Model 1: Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_acc = accuracy_score(y_test, lr.predict(X_test))

# Model 2: Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf.predict(X_test))

# Model 3: Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb.fit(X_train, y_train)
gb_acc = accuracy_score(y_test, gb.predict(X_test))

print(f"📊 Logistic Regression: {lr_acc*100:.2f}%")
print(f"🌲 Random Forest:       {rf_acc*100:.2f}%")
print(f"🚀 Gradient Boosting:   {gb_acc*100:.2f}%")
```

```python
# Best model select karo
best_pred = gb.predict(X_test)

print("🏆 Best Model: Gradient Boosting")
print("\nClassification Report:")
print(classification_report(y_test, best_pred,
      target_names=['Healthy', 'Diabetic']))

# Confusion Matrix
plt.figure(figsize=(6,4))
cm = confusion_matrix(y_test, best_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds',
            xticklabels=['Healthy','Diabetic'],
            yticklabels=['Healthy','Diabetic'])
plt.title('🩺 Diabetes - Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x='Importance', y='Feature',
            data=feature_importance, palette='Reds_r')
plt.title('🔑 Most Important Features for Diabetes')
plt.show()

print(feature_importance)
```

```python
new_patient = pd.DataFrame([{
    'preg': 2,      # Pregnancies
    'plas': 120,    # Glucose
    'pres': 70,     # Blood Pressure
    'skin': 25,     # Skin Thickness
    'insu': 80,     # Insulin
    'mass': 28.5,   # BMI
    'pedi': 0.5,    # Diabetes Pedigree
    'age': 35       # Age
}], columns=X.columns)

new_patient_scaled = scaler.transform(new_patient)
prediction = gb.predict(new_patient_scaled)
probability = gb.predict_proba(new_patient_scaled)

result = "🔴 Diabetic!" if prediction[0] == 1 else "✅ Healthy!"
print(f"Patient Result: {result}")
print(f"Probability: {probability[0][1]*100:.1f}% chance of diabetes")
```
