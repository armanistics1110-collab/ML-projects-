```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.datasets import fetch_openml

weather = fetch_openml(name='weather', version=1, as_frame=True)
df = weather.frame

print("Dataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nPehli 5 rows:")
df.head()
```

```python
# Target encode karo
le = LabelEncoder()
df['play'] = le.fit_transform(df['play'])

# Categorical columns encode karo
for col in df.select_dtypes(include='category').columns:
    df[col] = le.fit_transform(df[col])

df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()

print("Clean dataset shape:", df.shape)
print("\nTarget values:", df['play'].value_counts())
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Play count
df['play'].value_counts().plot(
    kind='bar', ax=axes[0],
    color=['#e74c3c', '#2ecc71']
)
axes[0].set_title('Weather Play Count\n(0=No Play, 1=Play)')

# Outlook distribution
df['outlook'].value_counts().plot(
    kind='pie', ax=axes[1],
    autopct='%1.1f%%',
    colors=['#3498db', '#f39c12', '#2ecc71']
)
axes[1].set_title('Outlook Distribution')

plt.tight_layout()
plt.show()
```

```python
X = df.drop('play', axis=1)
y = df['play']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# Model 1: Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf.predict(X_test))

# Model 2: Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb.fit(X_train, y_train)
gb_acc = accuracy_score(y_test, gb.predict(X_test))

print(f"🌲 Random Forest:     {rf_acc*100:.2f}%")
print(f"🚀 Gradient Boosting: {gb_acc*100:.2f}%")
```

```python
best_pred = rf.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, best_pred,
      target_names=['No Play', 'Play']))

plt.figure(figsize=(6,4))
cm = confusion_matrix(y_test, best_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Play','Play'],
            yticklabels=['No Play','Play'])
plt.title('🌦️ Weather - Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x='Importance', y='Feature',
            data=feature_importance, palette='Blues_r')
plt.title('🔑 Most Important Weather Features')
plt.show()
```

```python
new_weather = pd.DataFrame([{
    'outlook': 1,      # 0=Overcast, 1=Rainy, 2=Sunny
    'temperature': 75, # Temperature
    'humidity': 80,    # Humidity
    'windy': 0         # 0=No Wind, 1=Windy
}], columns=X.columns)

new_weather_scaled = scaler.transform(new_weather)
prediction = rf.predict(new_weather_scaled)

result = "⛳ Yes! Go Play!" if prediction[0] == 1 else "🏠 Stay Home!"
print(f"Weather Decision: {result}")
```
